```python
def multiply_A_to_A(A):
    n = len(A)
    AA = [[0 for _ in range(n)] for _ in range(n)]

    for i in range(n):
        for j in range(n):
            AA[i][j] = sum(A[i][k] * A[k][j] for k in range(n))

    return AA


A = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9],
]

A
multiply_A_to_A(A)
multiply_A_to_A(multiply_A_to_A(A))
multiply_A_to_A(multiply_A_to_A(multiply_A_to_A(A)))
```

```python
def multiply_A_to_A(A):
    n = len(A)
    AA = [[0 for _ in range(n)] for _ in range(n)]

    for i in range(n):
        for j in range(n):
            ith_row_of_A = A[i]
            jth_column_of_A = [row[j] for row in A]
            AA[i][j] = sum(ith_row_of_A[k] * jth_column_of_A[k]
                           for k in range(n))

    return AA


A = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9],
]

A
multiply_A_to_A(A)
multiply_A_to_A(multiply_A_to_A(A))
multiply_A_to_A(multiply_A_to_A(multiply_A_to_A(A)))
```

```python
def power(A, m):
    if m == 1:
        return A

    return power(multiply_A_to_A(A), m - 1)


power(A, 1)
power(A, 2)
power(A, 3)
power(A, 4)
```

```python
locker = [1, 9, 10, 5, 8, 13]
```

```python
i = 1
locker[i]
locker[:i] + locker[i + 1:]
```

```python
def check_locker(locker, needed):
    options = []
    for i in range(len(locker)):
        if locker[i] == needed:
            return True
        options.append((locker[:i] + locker[i + 1:], needed - locker[i]))

    return options


check_locker(locker, 5)
check_locker(locker, 26)
```
