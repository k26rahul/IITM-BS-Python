## multiply_matrices: Using the first approach

```python
# Function to multiply two square matrices
def multiply_matrices(A, B):
    # Get the size of the matrices
    n = len(A)

    # Initialize the result matrix with zeros
    AB = [[0 for _ in range(n)] for _ in range(n)]

    # Perform matrix multiplication
    for i in range(n):
        for j in range(n):
            AB[i][j] = sum(A[i][k] * B[k][j] for k in range(n))

    # Return the result matrix
    return AB


# Example matrix A
A = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9],
]

# Example usage of matrix multiplication
multiply_matrices(A, A)  # Multiply A by itself
multiply_matrices(A, multiply_matrices(A, A))  # Multiply A by A^2
```

## multiply_matrices: Using the second approach (More declarative and less error prone)

```python
# Function to multiply two square matrices using a different approach
def multiply_matrices(A, B):
    # Get the size of the matrices
    n = len(A)

    # Initialize the result matrix with zeros
    AB = [[0 for _ in range(n)] for _ in range(n)]

    # Perform matrix multiplication using row and column vectors
    for i in range(n):
        for j in range(n):
            # Extract ith row of matrix A
            ith_row_of_A = A[i]

            # Extract jth column of matrix B
            jth_column_of_B = [row[j] for row in B]

            # Calculate the dot product of the row and column vectors
            AB[i][j] = sum(ith_row_of_A[k] * jth_column_of_B[k]
                           for k in range(n))

    # Return the result matrix
    return AB


# Example matrix A
A = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9],
]

# Example usage of matrix multiplication
multiply_matrices(A, A)  # Multiply A by itself
multiply_matrices(A, multiply_matrices(A, A))  # Multiply A by A^2
```

## Final recursive function

```python
# Function to calculate the power of a square matrix
def power(A, m):
    # Base case: return the matrix itself for power 1
    if m == 1:
        return A

    # Recursive case: multiply the matrix by its (m-1) power
    return multiply_matrices(A, power(A, m - 1))


# Example matrix A
A = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9],
]

# Example usage of matrix power calculation
power(A, 1)  # A^1
power(A, 2)  # A^2
power(A, 3)  # A^3
power(A, 4)  # A^4
```

## Coins and locker problem

```python
locker = [1, 9, 10, 5, 8, 13]
```

```python
i = 1
locker[i]
locker[:i] + locker[i + 1:]
```

```python
def check_locker(locker, needed):
    options = []
    for i in range(len(locker)):
        if locker[i] == needed:
            return True
        options.append((locker[:i] + locker[i + 1:], needed - locker[i]))

    return options


check_locker(locker, 5)
check_locker(locker, 26)
```
